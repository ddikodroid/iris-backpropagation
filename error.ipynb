```python
import matplotlib.pyplot as plt
import csv
```

```python
x = []
y = []

with open('error.csv', 'r') as csvfile:
    plot = csv.reader(csvfile, delimiter=',')
    for row in plot:
        x.append(int(row[0]))
        y.append(float(row[1]))
        
plt.plot(x, y)

plt.title('Error Graph')
plt.xlabel('Epoch')
plt.ylabel('Error')

plt.show()
```
